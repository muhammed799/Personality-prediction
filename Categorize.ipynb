```python
def determine_baseline_angle(raw_baseline_angle):
    cat_type = ""
    if(raw_baseline_angle >= 0.2):
        baseline_angle = 0
        cat_type = "DESCENDING"
    elif(raw_baseline_angle <= -0.3):
        baseline_angle = 1
        cat_type = "ASCENDING"
    else:
        baseline_angle = 2
        cat_type = "STRAIGHT"
    
    return [baseline_angle, cat_type]
```

```python
def determine_top_margin(raw_top_margin):
    cat_type = ""
    if(raw_top_margin >= 1.7):
        top_margin = 0
        cat_type = "MEDIUM OR BIGGER"
    else:
        top_margin = 1
        cat_type = "NARROW"

    return top_margin, cat_type
```

```python
def determine_letter_size(raw_letter_size):
    cat_type = ""
    if(raw_letter_size >= 18.0):
        letter_size = 0
        cat_type = "BIG"
    elif(raw_letter_size < 13.0):
        letter_size = 1
        cat_type = "SMALL"
    else:
        letter_size = 2
        cat_type = "MEDIUM"

    return letter_size, cat_type
```

```python
def determine_line_spacing(raw_line_spacing):
    cat_type = ""
    if(raw_line_spacing >= 3.5):
        line_spacing = 0
        cat_type = "BIG"
    elif(raw_line_spacing < 2.0):
        line_spacing = 1
        cat_type = "SMALL"
    else:
        line_spacing = 2
        cat_type = "MEDIUM"

    return line_spacing, cat_type
```

```python
def determine_word_spacing(raw_word_spacing):
    cat_type = ""
    if(raw_word_spacing > 2.0):
        word_spacing = 0
        cat_type = "BIG"
    elif(raw_word_spacing < 1.2):
        word_spacing = 1
        cat_type = "SMALL"
    else:
        word_spacing = 2
        cat_type = "MEDIUM"

    return word_spacing, cat_type
```

```python
def determine_pen_pressure(raw_pen_pressure):
    cat_type = ""
    if(raw_pen_pressure > 180.0):
        pen_pressure = 0
        cat_type = "HEAVY"
    elif(raw_pen_pressure < 151.0):
        pen_pressure = 1
        cat_type = "LIGHT"
    else:
        pen_pressure = 2
        cat_type = "MEDIUM"

    return pen_pressure, cat_type
```

```python
def determine_slant_angle(raw_slant_angle):
    cat_type = ""
    if(raw_slant_angle == -45.0 or raw_slant_angle == -30.0):
        slant_angle = 0
        cat_type = "EXTREMELY RECLINED"
    elif(raw_slant_angle == -15.0 or raw_slant_angle == -5.0 ):
        slant_angle = 1
        cat_type = "A LITTLE OR MODERATELY RECLINED"
    elif(raw_slant_angle == 5.0 or raw_slant_angle == 15.0 ):
        slant_angle = 2
        cat_type = "A LITTLE INCLINED"
    elif(raw_slant_angle == 30.0 ):
        slant_angle = 3
        cat_type = "MODERATELY INCLINED"
    elif(raw_slant_angle == 45.0 ):
        slant_angle = 4
        cat_type = "EXTREMELY INCLINED"
    elif(raw_slant_angle == 0.0 ):
        slant_angle = 5
        cat_type = "STRAIGHT"
    else:
        slant_angle = 6
        cat_type = "IRREGULAR"

    return slant_angle, cat_type
```
